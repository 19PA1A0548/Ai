```
# importing seaborn
import seaborn as sns
```

```
#loading dataset using seaborn
df = sns.load_dataset("tips")
df.head()
```

```
#plotting using relplot
sns.relplot(x="total_bill",y="tip",hue="smoker",col="time",data=df,size="size")
```

```
sns.relplot(x="total_bill",y="tip",hue="smoker",col="time",kind="line",data=df,style="smoker")
```

```
sns.lmplot(x="total_bill",y="tip",col="time",hue="smoker",data=df)
#hue=for dividing the data into columns
#col=for dividing the graph no.of different values in it is no.of graphs
#kind= for plotting the graph
#x=the values in xaxis
#y=the values in yaxis
#size=used for asssigning the size of the point
```

```
#violin plot
sns.catplot(x="day",y="total_bill",data=df,kind="violin",hue="smoker",split=True)
```
